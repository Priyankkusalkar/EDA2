```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Load the dataset
data = pd.read_csv(r"C:\Users\Laptop\Downloads\data science Assignments\EDA2\EDA2\adult_with_headers.csv")
# Conduct basic data exploration
print("Summary Statistics:")
print(data.describe(include='all'))

print("\nData Types:")
print(data.dtypes)

print("\nMissing Values:")
print(data.isnull().sum())

# Handle missing values (imputation/removal)
# For simplicity, we'll drop rows with missing values here
# In practice, you might use imputation techniques
data = data.dropna()

# Verify missing values are handled
print("\nMissing Values After Handling:")
print(data.isnull().sum())

# Select numerical features for scaling
numerical_features = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']

# Apply Standard Scaling
standard_scaler = StandardScaler()
data_standard_scaled = data.copy()
data_standard_scaled[numerical_features] = standard_scaler.fit_transform(data[numerical_features])

# Apply Min-Max Scaling
min_max_scaler = MinMaxScaler()
data_min_max_scaled = data.copy()
data_min_max_scaled[numerical_features] = min_max_scaler.fit_transform(data[numerical_features])

# Output the scaled data for verification
print("\nStandard Scaled Data:")
print(data_standard_scaled[numerical_features].head())

print("\nMin-Max Scaled Data:")
print(data_min_max_scaled[numerical_features].head())

# Discussion on scaling techniques
"""
Standard Scaling:
- Transforms the data to have a mean of 0 and a standard deviation of 1.
- Preferred when the distribution of the numerical feature is approximately normal.
- Useful for algorithms that assume the data is normally distributed or sensitive to the magnitude of the data, such as linear regression, logistic regression, and SVM.

Min-Max Scaling:
- Transforms the data to be within a specified range, typically [0, 1].
- Preferred when the data does not follow a normal distribution or when preserving the relationships of the original data is important.
- Useful for algorithms that do not make assumptions about the data distribution, such as neural networks and k-nearest neighbors.

Choosing the right scaling technique depends on the specific requirements of your machine learning model and the nature of your data.
"""
```

#### . Encoding Techniques:
•	Apply One-Hot Encoding to categorical variables with less than 5 categories.
•	Use Label Encoding for categorical variables with more than 5 categories.
•	Discuss the pros and cons of One-Hot Encoding and Label Encoding


```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Identify categorical columns
categorical_columns = data.select_dtypes(include=['object']).columns

# Apply One-Hot Encoding to categorical variables with less than 5 categories
one_hot_columns = [col for col in categorical_columns if data[col].nunique() < 5]
data_one_hot_encoded = data.copy()
one_hot_encoder = OneHotEncoder(drop='first', sparse=False)
encoded_one_hot = one_hot_encoder.fit_transform(data[one_hot_columns])
encoded_one_hot_df = pd.DataFrame(encoded_one_hot, columns=one_hot_encoder.get_feature_names_out(one_hot_columns))
data_one_hot_encoded = data_one_hot_encoded.drop(one_hot_columns, axis=1)
data_one_hot_encoded = pd.concat([data_one_hot_encoded, encoded_one_hot_df], axis=1)

# Apply Label Encoding to categorical variables with 5 or more categories
label_columns = [col for col in categorical_columns if data[col].nunique() >= 5]
data_label_encoded = data.copy()
label_encoders = {}
for col in label_columns:
    label_encoders[col] = LabelEncoder()
    data_label_encoded[col] = label_encoders[col].fit_transform(data[col])

# Output the encoded data for verification
print("\nOne-Hot Encoded Data:")
print(data_one_hot_encoded.head())

print("\nLabel Encoded Data:")
print(data_label_encoded.head())

# Discussion on encoding techniques
"""
One-Hot Encoding:
Pros:
- Avoids the introduction of ordinal relationships among categories.
- Preferred for nominal categorical variables with no intrinsic order.
- Makes the data suitable for many machine learning algorithms that expect numerical input without order bias.

Cons:
- Can lead to high-dimensional data, especially when there are many categories.
- Increases the computational cost and complexity, particularly for models sensitive to the curse of dimensionality (e.g., tree-based models).

Label Encoding:
Pros:
- Simple and easy to implement.
- Does not increase the dimensionality of the dataset.
- Useful for ordinal categorical variables where the categories have a meaningful order.

Cons:
- Introduces ordinal relationships among categories even when there are none, which can mislead some algorithms.
- Not suitable for nominal variables as it may introduce a bias towards categories with higher numerical values.

Choosing between One-Hot Encoding and Label Encoding depends on the nature of the categorical variables (nominal vs. ordinal) and the requirements of the machine learning algorithm being used.
"""
```
